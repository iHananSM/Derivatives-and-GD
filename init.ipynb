```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def diff(f, x, h=1e-5):
    return (f(x+h) - f(x))/h
```

```python
init_value = np.random.randint(50)
init_value = x = init_value
step_rate = .1
steps = epochs = 20
```

```python
x_vals_plot = np.linspace(-100, 100, 200)
y_vals_plot = f(x_vals_plot)
```

```python
plt.figure(figsize=(8, 5))
for i in range(epochs):
    grad = diff(f, x)
    x = x - step_rate*grad

    plt.clf()
    plt.plot(x_vals_plot, y_vals_plot, label="f(x) = x", color="blue")
    
    plt.scatter([x], [f(x)], color="red", zorder=5)
    plt.title(f"Step {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.pause(0.3)  # Pause to simulate animation

plt.show()
```
